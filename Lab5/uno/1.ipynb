```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams.update({
    "text.usetex": False,
    "font.family": "serif",
    "font.size": 12,
    "axes.grid": True,
    "grid.alpha": 0.5,
    "lines.linewidth": 2
})

t = np.linspace(-1.5, 1.5, 1000)
pi_t = np.where(np.abs(t) <= 0.5, 1, 0)

plt.figure(figsize=(8, 4))
plt.plot(t, pi_t, color='#282CFF')
plt.ylim(-0.2, 1.2)
plt.xlabel('t')
plt.ylabel(r'$\Pi(t)$')
plt.tight_layout()
plt.savefig("pi_t.pdf", format="pdf")
plt.show()

nu = np.linspace(-5, 5, 1000)
pi_hat = np.sinc(nu)

plt.figure(figsize=(8, 4))
plt.plot(nu, pi_hat, color='#FF282C')
plt.axhline(0, color='black', linewidth=1)
plt.xlabel(r'$\nu$')
plt.ylabel(r'$\hat{\Pi}(\nu)$')
plt.tight_layout()
plt.savefig("pi_hat.pdf", format="pdf")
plt.show()
```
